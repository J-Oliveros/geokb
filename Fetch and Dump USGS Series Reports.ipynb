There are over 80K USGS Numbered Series reports in the PW Catalog. In order to get a full handle on the available information and to establish a baseline in the GeoKB, I run this to build a dump of the raw web service response. We can then operate on that to tease out the parts we need. After we get a baseline representation for all identifiers in the knowledgebase, we can run regular updates using a parameter (mod_x_days) based on the last modified date.

```python
import requests
import pickle
```

```python
%%time
records = []
page_num = 1

while True:
    api = f"https://pubs.er.usgs.gov/pubs-services/publication/?page_size=1000&page_number={page_num}&subtypeName=USGS%20Numbered%20Series"
    pw = requests.get(api)
    
    if pw.status_code != 200:
        break

    pw_records = pw.json()
    
    if pw_records["records"]:
        records.extend(pw_records["records"])
        page_num += 1
        print(page_num, len(records))
    else:
        break

pickle.dump(records, open('data/pw_usgs_reports_dump.pickle', 'wb'))
```
